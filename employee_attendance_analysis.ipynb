```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('employee_attendance.csv')
df.head()
```

Calculate attendance percentage

```python
df["attendance"] = (df["Days_Present"] / df["Total_Working_Days"]) *100
print(df[["Name","attendance"]])
```

Average attendance by department

```python
avg_attendance = df.groupby("Department")["attendance"].mean().reset_index()
print(avg_attendance)
```

```python
top_emp = df.loc[df["attendance"].idxmax()]
print(top_emp)
```

top employee

```python
plt.figure(figsize=(10,6))
plt.bar(df["Name"], df["attendance"],color="skyblue")
plt.title("Employee Attendance Percentage (January)")
plt.xlabel("Employees")
plt.ylabel("Attendance (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(avg_attendance["Department"], avg_attendance["attendance"], color='orange')
plt.title("Average Attendance by Department")
plt.xlabel("Department")
plt.ylabel("Average Attendance (%)")
plt.tight_layout()
plt.show()
```
